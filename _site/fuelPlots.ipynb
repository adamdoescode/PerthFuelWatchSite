```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from plotnine import ggplot, geom_point, aes, stat_smooth, facet_wrap, scale_x_continuous, theme_bw, ggtitle, geom_col, scale_x_discrete, theme, element_text, after_stat
from datetime import datetime
```

# Fuel plots to explore the data!

A place to explore the data bit ðŸ˜‰

```python
TODAYS_DATE = datetime.today().strftime("%d-%m-%Y")
#import data
pricesPerth = pd.read_csv(f'data/{TODAYS_DATE}-pricesPerth.csv')
pricesPerth
```

```python
plotNOR_SOR = (ggplot(pricesPerth, aes(x='latitude', y='price', color='brand'))
    + geom_point()
    + ggtitle("Prices vs latitude coloured by brand")
    )
plotNOR_SOR
```

```python
meanBrandPrice = pricesPerth.groupby('brand').mean()['price']
```

```python
brandPriceAgg = pricesPerth.groupby('brand').agg({'price': ['mean', 'median', 'std']})
brandPriceAgg.reset_index(inplace=True, drop=True)
brandPriceAgg
```

```python
brandsBarPlot = (
    ggplot(pricesPerth, aes(x="brand", y="price")) +
    geom_col(aes(after_stat('mean'))) +
    theme(axis_text_x=element_text(rotation=90, hjust=1))
)
brandsBarPlot
```
